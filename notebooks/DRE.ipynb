# Receita Bruta


```python
# 1) Importar as bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', 20)

# 2) Importar CSV
receitas = pd.read_csv('../data/Contas A Receber detalhado por Data de Vencimento.csv', sep=";")


def convert_currency(value):
    value = value.replace('R$', '').replace('.', '').replace(',', '.').strip()
    return float(value)
receitas["Vencimento"] = pd.to_datetime(receitas['Vencimento'], format='%d/%m/%Y' )
receitas["Valor"] = receitas["Valor"].apply(convert_currency)

receitas.head()
```

```python
receitasPorTipo = receitas.groupby("Descrição").agg({"Valor": sum}).sort_values(by = "Valor", ascending=False)
receitasPorTipo["percent"] = receitasPorTipo["Valor"] / receitasPorTipo["Valor"].sum()
receitasPorTipo["cum"] = (receitasPorTipo["percent"].cumsum())


receitasPorTipo
```

```python
receitasPorStatus = receitas.groupby(by="Status do Débito").agg({"Valor":sum})
receitasPorStatus["percent"] = (receitasPorStatus["Valor"]/receitasPorStatus["Valor"].sum())

receitasPorStatus
```

```python
receitasTotais =  receitas[receitas["Status do Débito"] == "Pago"]["Valor"].sum()
print(receitasTotais)
```

# DESPESAS

```python
despesas = pd.read_csv("../data/Contas A Pagar por Data de Vencimento.csv", sep = ";")

despesas["Vencimento"] = pd.to_datetime(despesas['Vencimento'], format='%d/%m/%Y')
despesas["Valor"] = despesas["Valor"].apply(convert_currency)



despesas.head()
```

```python
despesasPorTipo = despesas.groupby(by="Natureza de Operação").agg({"Valor":sum}).sort_values(by="Valor", ascending=False)
despesasPorTipo["percent"] = despesasPorTipo["Valor"]/despesasPorTipo["Valor"].sum()
despesasPorTipo["cum"] = despesasPorTipo["percent"].cumsum()



despesasPorTipo
```

```python
despesasTotais = -1*despesas["Valor"].sum()
print(despesasTotais)
```

# EBTIDA

```python
df1 = pd.DataFrame([receitasTotais, despesasTotais, receitasTotais+despesasTotais], index=["RECEITAS", "CUSTOS E DESPESAS", "EBTIDA"], columns=["DRE"])
df1
df1.to_csv('out.csv', sep=";")
```
