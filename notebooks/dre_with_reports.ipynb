# Receita Bruta


```python
# 1) Importar as bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', 20)

# 2) Importar CSV
receitas = pd.read_csv('../data/Relatório de recebimentos por Natureza de Operação dia a dia.csv', sep=";")


def convert_currency(value):
    value = value.replace('R$', '').replace('.', '').replace(',', '.').strip()
    return float(value)
receitas["Valor"] = round(receitas["Valor"].apply(convert_currency),2)

receitas.head()
```

```python
receitasPorTipo = receitas.groupby("Natureza de Operação").agg({"Valor": sum}).sort_values(by = "Valor", ascending=False)
receitasPorTipo["percent"] = round(receitasPorTipo["Valor"] / receitasPorTipo["Valor"].sum(), 2)
receitasPorTipo["cum"] = (receitasPorTipo["percent"].cumsum())

receitasPorTipo.to_csv('report_receitas.csv', sep=";")

receitasPorTipo

```

```python
receitasTotais =  receitas["Valor"].sum()
print(receitasTotais)
```

# DESPESAS

```python
despesas = pd.read_csv("../data/Resumo de Pagamentos por Natureza de Operação.csv", sep = ";")

despesas["Valor"] = round(despesas["Valor"].apply(convert_currency),2)



despesas.head()
```

```python
despesasPorTipo = despesas.groupby(by="Natureza de Operação").agg({"Valor":sum}).sort_values(by="Valor", ascending=False)
despesasPorTipo["percent"] = round(despesasPorTipo["Valor"]/despesasPorTipo["Valor"].sum(),2)
despesasPorTipo["cum"] = despesasPorTipo["percent"].cumsum()


despesasPorTipo.to_csv('report_despesas.csv', sep=";")

despesasPorTipo
```

```python
despesasTotais = -1*despesas["Valor"].sum()
print(despesasTotais)
```

# EBTIDA

```python
df1 = pd.DataFrame([receitasTotais, despesasTotais, receitasTotais+despesasTotais], index=["RECEITAS", "CUSTOS E DESPESAS", "EBTIDA"], columns=["DRE"])
df1.to_csv('report_dre.csv', sep=";")

df1
```
